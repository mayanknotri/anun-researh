```python
from openpyxl import load_workbook
import pandas as dn
```

```python
wb = load_workbook(filename='pelanggan12.xlsx')
sheet_range = wb['Sheet1']

df = dn.DataFrame (sheet_range.values)
df
```

```python
x = df[1:8] [[1,2,3,4,5,6,7,8]]
x.columns = ['tanggal','nama','alamat','no_telp','nama_barang','jumlah','harga','jumlahuang']
x
```

## jumlah barang

```python
for index, row in x.iterrows():
    x.ix[index, 'jumlahbarang']=row['jumlah'][-5]
x
```

## total harga

```python
for index, row in x.iterrows():
    x.ix[index, 'total harga']=int(row['jumlahbarang'])*row['harga']
x
```

## diskon

```python
for index, row in x.iterrows():
    x.ix[index, 'diskon']=row['total harga']*(row['total harga']*15/100)
x
```

## kembalian

```python
for index, row in x.iterrows():
    x.ix[index, 'kembalian']=row['diskon']-row['jumlahuang']
x    
```
